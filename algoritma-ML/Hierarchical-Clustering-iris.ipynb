```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
```

```python
# Langkah 1: Memuat dataset Iris
iris = load_iris()
data = iris.data  # Menggunakan fitur: [sepal length, sepal width, petal length, petal width]

# Langkah 2: Standarisasi data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Langkah 3: Menghitung linkage matrix
linkage_matrix = linkage(data_scaled, method='ward')

# Langkah 4: Visualisasi dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, truncate_mode='lastp', p=30, leaf_font_size=10)
plt.title("Dendrogram - Hierarchical Clustering on Iris Dataset")
plt.xlabel("Sample Index or (Cluster Size)")
plt.ylabel("Distance")
plt.show()
```

```python
# Langkah 5: Membagi data menjadi 3 cluster
clusters = fcluster(linkage_matrix, t=3, criterion='maxclust')

# Menampilkan hasil cluster
print("Cluster assignments for Iris data:", clusters)
```

```python
# Langkah 6: Menambahkan informasi cluster ke dataset
import pandas as pd
df = pd.DataFrame(data, columns=iris.feature_names)
df['Cluster'] = clusters

# Menampilkan ringkasan cluster
print(df.groupby('Cluster').mean())
```
