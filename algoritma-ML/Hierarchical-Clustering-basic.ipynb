```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
```

```python
# Contoh data dummy (2D data points)
data = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [8, 7],
    [9, 8],
    [10, 8]
])

# Langkah 1: Menghitung linkage matrix menggunakan metode 'ward'
# (Ward meminimalkan varian antar cluster)
linkage_matrix = linkage(data, method='ward')

# Langkah 2: Membuat dendrogram
plt.figure(figsize=(8, 5))
dendrogram(linkage_matrix, labels=np.arange(1, len(data)+1), leaf_font_size=10)
plt.title("Dendrogram - Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# Membagi data menjadi 2 cluster berdasarkan jarak maksimal
clusters = fcluster(linkage_matrix, t=2, criterion='maxclust')

# Output cluster untuk setiap data
print("Cluster assignments:", clusters)
```
