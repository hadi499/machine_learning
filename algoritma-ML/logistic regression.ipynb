```python
# Import library
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 1. Load dataset
iris = load_iris()

# Menggunakan hanya dua kelas (Binary classification)
import pandas as pd
X = iris.data
y = (iris.target == 2).astype(int)  # Menjadikan target binary (Iris-Virginica == 1, lainnya == 0)

# 2. Split data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Inisialisasi Logistic Regression dan training
model = LogisticRegression()
model.fit(X_train, y_train)

# 4. Prediksi
y_pred = model.predict(X_test)
```

```python
# 5. Evaluasi
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# 6. Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ambil satu fitur untuk visualisasi
X_feature = X[:, 0].reshape(-1, 1)  # Fitur pertama (sepal length)
model = LogisticRegression()
model.fit(X_feature, y)  # Latih model

# Plot sigmoid curve
X_range = np.linspace(X_feature.min(), X_feature.max(), 300).reshape(-1, 1)
probabilities = model.predict_proba(X_range)[:, 1]  # Probabilitas untuk kelas 1

plt.figure(figsize=(8, 6))
plt.scatter(X_feature, y, c='blue', label='Data', alpha=0.7)
plt.plot(X_range, probabilities, color='red', label='Sigmoid Curve')
plt.title("Logistic Regression - Sigmoid Curve")
plt.xlabel("Sepal Length")
plt.ylabel("Probability of being Iris-Virginica")
plt.legend()
plt.show()
```
